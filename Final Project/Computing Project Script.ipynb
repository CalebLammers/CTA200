```python
def create_plot(x, y, x_lim, x_label, y_lim, y_label, loglog=True):
    """Creates plot of x vs. y with xlim of x_lim and ylim of y_lim and
    respective labels (log-log scale or log-y scale)

    Parameters:
    x - np array of x values
    y - np array of corresponding y values
    x_lim - Tuple containing leftmost and rightmost x value to be on graph
    y_lim - Tuple containing lowest and highest y value to be on graph
    x_label - Label for x-axis
    y_label - Label for y-axis
    loglog - By default, graphs with a log x and y axis, if true, graphs
    with a log y axis only
    

    Returns:
    Function has no return value, however, it shows the matplotlib graph
    """
    plt.figure(figsize=(15, 9))
    
    if loglog:
        plt.loglog(x, y)
    else:
        plt.semilogy(x, y)
        
    plt.xlabel(x_label, fontsize=14)
    plt.xlim(x_lim)
    plt.ylabel(y_label, fontsize=14)
    plt.ylim(y_lim)
    plt.savefig("plot.pdf")
    plt.show()
```

```python
def linear_interpolation(x_s, y_s, x_val):
    """Determines y value at x_val by linearly interpolating y_s

    Parameters:
    x_s - np array of x values
    y_s - np array of corresponding y values
    x_val - x value at which y_s will be linearly interpolated
    

    Returns:
    y value linearly interpolated from y_s at x_val
    """
    index = np.abs(x_s - x_val).argmin()
    
    if x_s[index] > x_val:
        l_index = index - 1
        r_index = index
    elif x_s[index] < x_val:
        l_index = index
        r_index = index + 1
    else:
        return y_s[index]
        
        
    m = (y_s[r_index] - y_s[l_index])/(x_s[r_index] - x_s[l_index])
    b = y_s[r_index] - m*x_s[r_index]
    
    return b + m*x_val
```

```python
def get_spectral_density(wave, spec, z):
    """Converts spectrum from (bolometric) luminosity in units of solar luminosities 
    per hertz to flux density in units of microJansky

    Parameters:
    wave - np array of wavelengths (units don't matter)
    spec - np array of luminosities measured in solar luminosities per hertz
    z - Redshift of the galaxy

    Returns:
    New np array of wavelengths and the corresponding np array of flux densities
    measured in microJansky
    """
    d = cosmo.luminosity_distance(z)
    spec = 4.02*(10**13)*((1 + z)*spec)/(4*math.pi*d.value**2)
    wave = wave*(1 + z)
    
    return wave, spec
```

```python
def read_file(name):
    """Creates 3 np arrays of data, read in from a file consisting of three columns

    Parameters:
    name - name of file

    Returns:
    3 np arrays containing data in the three coloumns from the file
    """
    lst_1 = []
    lst_2 = []
    lst_3 = []
    
    with open(name) as reader:
        for line in reader:
            list_str = line.split()
            if len(list_str) == 3:
                lst_1.append(float(list_str[0]))
                lst_2.append(float(list_str[1]))
                lst_3.append(float(list_str[2]))
    return np.array(lst_1), np.array(lst_2), np.array(lst_3)
```

```python
def weighted_fitting(x_s, y_s, errs):
    """Calculates y-int and slope (and uncertainties) of 
    line that fits data using weighted linear least-square fitting

    Parameters:
    x_s - numpy array of x values
    y_s - numpy array of corresponding y values
    errs - numpy array of corresponding errors in the y values

    Returns:
    Tuple containing y-int, uncertainty in y-int, slope, and 
    uncertainy in slope
    """
    list_Y = []
    list_A = []
    list_C = []
    
    for i in range(len(x_s)):
        list_Y.append([y_s[i]])
        list_A.append([1, x_s[i]])
        
        C_row = []
        
        counter = 0
        while counter < len(x_s):
            if counter == i:
                C_row.append(errs[i]**2)
            else:
                C_row.append(0)
                
            counter += 1
        
        list_C.append(C_row)
            
        
    A = Matrix(list_A)
    Y = Matrix(list_Y)
    C = Matrix(list_C)
    
    cov = ((A.T*C.inv()*A).inv())
    
    b_error = cov[0, 0]
    m_error = cov[1, 1]
    
    X = cov*(A.T*C.inv()*Y)
    
    return X[0, 0], sqrt(b_error), X[1, 0], sqrt(m_error)
```

```python
def calc_chi(x_s, y_s, errs, b, m):
    """Calculates chi^2, a measure of how closely a line fits
    the given data

    Parameters:
    x_s - np array of x values
    y_s - np array of corresponding y values
    errs - np array of corresponding errors in the y values
    b - y-int of line
    m - slope of line

    Returns:
    The value of chi^2
    """
    return np.sum(((y_s - (b + m*x_s))**2)/(errs**2))
```

```python
x_s, y_s, errs = read_file('fitting_simple_data.txt')
            
b, b_err, m, m_err = weighted_fitting(x_s, y_s, errs)

txt = "\mathrm{{{3}}} = {0:.3f}_{{-{1:.3f}}}^{{{2:.3f}}}"
txt = txt.format(b, b_err, b_err, 'b')
display(Math(txt))
txt = "\mathrm{{{3}}} = {0:.3f}_{{-{1:.3f}}}^{{{2:.3f}}}"
txt = txt.format(m, m_err, m_err, 'm')
display(Math(txt))

x_line = np.linspace(0, 12)
y_line = b + m*x_line       

plt.figure(figsize=(15, 9))
plt.scatter(x_s, y_s)
plt.errorbar(x_s, y_s, yerr=errs, fmt=".", capsize=0)
plt.xlabel('x')
plt.ylabel('y')

plt.plot(x_line, y_line)

plt.show()

print('chi^2 = ', calc_chi(x_s, y_s, errs, b, m))
```

```python
def spectrum_likelihood(wave_obs, spec_obs, errs, wave_theo, spec_theo):
    """Calculates the likelihood that a theoretical SED matches
    an observed SED

    Parameters:
    wave_obs - np array of observed wavelength values
    spec_obs - np array of corresponding observed flux values
    errs - np array of errors in observed flux values
    wave_theo - np array of theoretical wavelength values
    spec_theo - np array of corresponding theoretical flux values

    Returns:
    The likelihood, note that it is always less than 0
    """
    interpol_flux = []
    
    for i in range(len(wave_obs)):
        interpol_flux.append(linear_interpolation(wave_theo, spec_theo, wave_obs[i]))
                   
    inter_flux = np.array(interpol_flux)
    
    return -np.sum(((spec_obs - inter_flux)**2)/(errs**2))
```

```python
def find_optimal_z_age(z_start, z_incr, z_end, age_start, age_incr, age_end, name):
    """Uses a brute force loop to estimate age and z that
    maximize the likelihood function

    Parameters:
    z_start - Starting z (the loop will start with z_start + z_incr)
    z_incr - Increment in z
    z_end - Value of z the loop will stop at
    age_start - Starting age (the loop will start with age_start + age_incr)
    age_incr - Increment in age
    age_end - Value of age the loop will stop at
    name - Name of file

    Returns:
    Returns tuple of the age and z that maximized the likelihood function
    """
    wave_obs, spec_obs, errs = read_file(name)
    max_likelihood = -10**10
    max_z = -1
    max_age = -1
    z = z_start
    
    while z <= z_end:
        z += z_incr
        age = age_start + age_incr
        while age <= age_end:
            wave_theo, spec_theo = generate_SED(age, z)
            likelihood = round(spectrum_likelihood(wave_obs, spec_obs, errs, wave_theo, spec_theo), 3)
            
            if likelihood > max_likelihood:
                max_likelihood = likelihood
                max_z = round(z, 3)
                max_age = round(age, 3)
        
            age += age_incr
    
    return max_age, max_z
```

```python
def log_likelihood(theta, x, y, yerr):
    """Calculates the log of the likelihood that the values
    match the model

    Parameters:
    theta - Tuple containing slope, y-int, and f parameter
    x - np array of x values
    y - np array of corresponding y values
    yerr - np array of corresponding errors in the y values

    Returns:
    Returns the log of the likelihood that the y values
    fit the model
    """
    m, b, f = theta
    model = m*x + b
    sigma2 = yerr**2 + (model**2)*(f**2)
    return -0.5*np.sum(((y - model)**2)/sigma2 + np.log(sigma2))
```

```python
def create_rand_pos(sig_1, mu_1, sig_2, mu_2, num_walkers):
    """Creates random position list for the walkers

    Parameters:
    sig_1 - Standard deviation for first parameter
    mu_1 - Mean value of first parameter
    sig_2 - Standard deviation for second parameter
    mu_2 - Mean value of second parameter
    num_walkers - Number of walkers

    Returns:
    2D np array contaning initial position for the walkers
    """    
    array_1 = sig_1 * np.random.randn(num_walkers) + mu_1
    array_2 = sig_2 * np.random.randn(num_walkers) + mu_2
    
    lst_pos = []
    for i in range(len(array_1)):
        lst_pos.append([array_1[i], array_2[i]])
        
    return np.array(lst_pos)
```

```python
def prior(theta):
    """Calculates the prior probability that the values
    matches the model

    Parameters:
    theta - Tuple containing slope, y-int, and f parameter

    Returns:
    Returns the 1.0 if the parameters are within a
    reasonable range and 0.0 otherwise
    """
    m, b, f = theta
    if 2 < m < 12 and -15.0 < b < 15.0 and 0 < f < 100:
        return 1.0
    return 0.0
```

```python
def log_probability(theta, x, y, yerr):
    """Calculates the log of the probability that 
    the values matches the model

    Parameters:
    theta - Tuple containing slope, y-int, and f parameter
    x - np array of x values
    y - np array of corresponding y values
    yerr - np array of corresponding errors in the y values

    Returns:
    Returns the sum of the log of the likelihood and the 
    log of the prior
    """    
    lp = np.log(prior(theta))
    if not np.isfinite(lp):
        return -np.inf
    return lp + log_likelihood(theta, x, y, yerr)
```

```python
def galaxy_prior(theta):
    """Calculates the prior probability that the values
    matches the model

    Parameters:
    theta - Tuple containing log(age) and z

    Returns:
    Returns the 1.0 if the parameters are within a
    reasonable range and 0.0 otherwise
    """    
    log_age, z = theta
    if 0 < log_age < 5 and 0 < z < 2.0:
        return 1.0
    return 0.0
```

```python
def galaxy_probability(theta, wave_obs, spec_obs, errs):
    """Calculates the log of the probability that 
    the values matches the model

    Parameters:
    theta - Tuple containing log(age) and z
    wave_obs - np array of observed wavelength values
    spec_obs - np array of corresponding observed flux values
    errs - np array of corresponding errors in the flux values

    Returns:
    Returns the sum of the log of the likelihood and the 
    log of the prior
    """
    lp = np.log(galaxy_prior(theta))
    if not np.isfinite(lp):
        return -np.inf
    
    log_age, z = theta
    wave_theo, spec_theo = generate_SED(np.exp(log_age), z)
    return lp + spectrum_likelihood(wave_obs, spec_obs, errs, wave_theo, spec_theo)
```
